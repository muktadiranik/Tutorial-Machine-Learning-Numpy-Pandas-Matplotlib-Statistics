```python
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday
```

```python
class BDHolidayCalendar(AbstractHolidayCalendar):
    rules = [
        Holiday('Independence Day', month=3, day=26, observance=nearest_workday),
        Holiday('Victory Day', month=12, day=16)
    ]
BDCalendar = CustomBusinessDay(calendar=BDHolidayCalendar())
```

```python
pd.date_range(start='2016-12-01', end='2016-12-31', freq=BDCalendar)
```

```python
egypt_date = CustomBusinessDay(weekmask='Sun Mon Tue Wed Thu Fri')
```

```python
pd.date_range(start='2020-12-01', periods=30, freq=egypt_date)
```

```python
India_Date = CustomBusinessDay(weekmask='Mon Tue Wed Thu Fri', holidays=['2020-12-12'])
```

```python
pd.date_range(start='2020-12-01', periods=30, freq=India_Date)
```
