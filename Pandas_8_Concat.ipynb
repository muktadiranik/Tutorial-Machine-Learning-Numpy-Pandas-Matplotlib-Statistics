```python
import pandas as pd
```

```python
df = pd.read_csv('weather_by_cities.csv')
df
```

```python
df_bd = pd.DataFrame({
    'city': ['rajshahi', 'rongpur', 'bogura'], 
    'temperature': [30, 32, 28], 
    'humidity': [75, 85, 95]
})
df_us = pd.DataFrame({
    'city': ['newyork', 'chicago', 'orlando'], 
    'temperature': [14, 16, 18], 
    'humidity': [70, 80, 90]
})
```

```python
df_compact = pd.concat([df_bd, df_us])
df_compact
```

```python
df_compact = pd.concat([df_bd, df_us], ignore_index=True)
df_compact
```

```python
df_compact = pd.concat([df_bd, df_us], keys=['bd', 'us'])
df_compact
```
