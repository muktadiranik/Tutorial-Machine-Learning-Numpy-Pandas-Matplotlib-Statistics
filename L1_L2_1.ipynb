```python
import pandas as pd
```

```python
df = pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
df.drop(['Address', 'Date', 'YearBuilt', 'Lattitude', 'Longtitude'], axis='columns', inplace=True)
```

```python
df.head()
```

```python
df.drop(['Method'], axis='columns', inplace=True)
```

```python
df.drop(['Postcode'], axis='columns', inplace=True)
```

```python
df.drop(['CouncilArea'], axis='columns', inplace=True)
```

```python
df.drop(['Suburb'], axis='columns', inplace=True)
```

```python
df.drop(['Regionname'], axis='columns', inplace=True)
```

```python
df.head()
```

```python
fillna = ['Bedroom2', 'Bathroom', 'Car']
```

```python
df[fillna] = df[fillna].fillna(0)
df.isna().sum()
```

```python
df.shape
```

```python
df['Landsize'] = df['Landsize'].fillna(df['Landsize'].mean())
```

```python
df.isna().sum()
```

```python
df['BuildingArea'] = df['BuildingArea'].fillna(df['BuildingArea'].mean())
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.head()
```

```python
x = df.drop(['Price'], axis=1)
```

```python
x
```

```python
y = df['Price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
(xtrain, xtest, ytrain, ytest) = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(xtrain, ytrain)
```

```python
lr.score(xtest, ytest)
```

```python
from sklearn.linear_model import Lasso
```

```python
lsr = Lasso(alpha=5, max_iter=10000000, tol=0.0001)
```

```python
lsr.fit(xtrain, ytrain)
```

```python
lsr.score(xtest, ytest)
```

```python
from sklearn.linear_model import Ridge
```

```python
rdr = Ridge(alpha=5, max_iter=10000000, tol=0.0001)
```

```python
rdr.fit(xtrain, ytrain)
```

```python
rdr.score(xtest, ytest)
```
