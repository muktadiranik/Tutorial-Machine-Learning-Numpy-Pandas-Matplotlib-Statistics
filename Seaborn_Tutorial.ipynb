Pair Grid

```python
iris = sb.load_dataset('iris')
iris_grid = sb.PairGrid(iris, hue='species')
iris_grid.map(plt.scatter)
```

```python
iris = sb.load_dataset('iris')
iris_grid = sb.PairGrid(iris, hue='species')
iris_grid.map(plt.scatter)
iris_grid.map_diag(plt.hist)
```

```python
iris = sb.load_dataset('iris')
iris_grid = sb.PairGrid(iris, hue='species')
iris_grid.map_offdiag(plt.scatter)
iris_grid.map_diag(plt.hist)
```

```python
iris = sb.load_dataset('iris')
iris_grid = sb.PairGrid(iris, hue='species')
iris_grid.map_diag(plt.hist)
iris_grid.map_upper(plt.scatter)
iris_grid.map_lower(sb.kdeplot)
```

```python
iris = sb.load_dataset('iris')
iris_grid = sb.PairGrid(iris, hue='species', 
                       x_vars=['sepal_length', 'sepal_width'], 
                       y_vars=['petal_length', 'petal_width'])
iris_grid.map(plt.scatter)
iris_grid.add_legend()
```

Facet Grid

```python
tips_df = sb.load_dataset('tips')
```

```python
tips_facet = sb.FacetGrid(tips_df, col='time', row='smoker')
tips_facet.map(plt.hist, 'total_bill', bins=8)
```

```python
tips_facet = sb.FacetGrid(tips_df, col='time', hue='smoker')
tips_facet.map(plt.scatter, 'total_bill', 'tip')
```

```python
tips_facet = sb.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.5)
tips_facet.map(plt.scatter, 'total_bill', 'tip')
```

```python
tips_facet = sb.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.5, 
                         col_order=['Dinner', 'Lunch'])
tips_facet.map(plt.scatter, 'total_bill', 'tip')
tips_facet.add_legend()
```
