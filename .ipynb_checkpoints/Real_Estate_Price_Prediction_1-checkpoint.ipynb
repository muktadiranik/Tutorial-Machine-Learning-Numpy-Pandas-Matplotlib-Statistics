```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
def convert_to_float(x):
    token = x.split('-')
    if len(token) == 2:
        return ((float(token[0]) + float(token[1])) / 2)
    try:
        return float(x)
    except:
        return None
```

```python
convert_to_float('2100 - 2850')
```

```python
exclude = np.array([])
for row, column in df.groupby('location'):
    bhk = {}
    for row, column in column.groupby('bhk'):
        bhk[row] = {
            'mean': column.price_per_sqft.mean(),
            'std':column.price_per_sqft.std(),
            'count': column.shape[0]
        }
    for row, column in column.groupby('bhk'):
        info = bhk.get(row - 1)
        if info and info['count'] > 5:
            exclude = np.append(exclude, column[column.price_per_sqft < (info['mean'])].index.values)
df.drop(exclude, axis='index', inplace=True)
```
