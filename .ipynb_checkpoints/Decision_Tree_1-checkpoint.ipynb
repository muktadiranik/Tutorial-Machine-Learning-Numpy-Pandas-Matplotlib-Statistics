```python
import pandas as pd
df = pd.read_csv('salaries.csv')
df
```

```python
com_dummies = pd.get_dummies(df['company'])
job_dummies = pd.get_dummies(df['job'])
degree_dummies = pd.get_dummies(df['degree'])
```

```python
com_dummies.drop(['abc pharma'], axis='columns', inplace=True)
```

```python
com_dummies
```

```python
job_dummies.drop(['business manager'], axis='columns', inplace=True)
```

```python
job_dummies
```

```python
degree_dummies.drop(['bachelors'], axis='columns', inplace=True)
```

```python
degree_dummies
```

```python
df.drop(['company', 'job', 'degree'], axis='columns', inplace=True)
```

```python
dummies = pd.concat([com_dummies, job_dummies, degree_dummies], axis='columns')
```

```python
dummies
```

```python
df = pd.concat([df, dummies], axis='columns')
```

```python
df
```

```python
x = df[['facebook', 'google', 'computer programmer', 'sales executive', 'masters']]
y = df['salary_more_then_100k']
```

```python
from sklearn import tree
dtc = tree.DecisionTreeClassifier()
```

```python
dtc.fit(x, y)
```

```python
dtc.predict([[0, 0, 0, 0, 0]])
```

```python
import seaborn as sb
```

```python
import matplotlib.pyplot as plt
```

```python
y_predicted = dtc.predict(x)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_predicted, y)
```

```python
cm
```

```python
plt.figure(figsize= (10, 8))
sb.heatmap(cm, annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
```
