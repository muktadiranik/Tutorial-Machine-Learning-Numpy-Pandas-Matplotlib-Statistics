```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_temp = [50, 51, 52, 48, 47, 49, 46]
min_temp = [43, 42, 40, 44, 33, 35, 37]
avg_temp = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.plot(days, max_temp)
```

```python
plt.plot(days, max_temp)
plt.plot(days, min_temp)
plt.plot(days, avg_temp)
```

```python
plt.plot(days, max_temp)
plt.plot(days, min_temp)
plt.plot(days, avg_temp)
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
```

```python
plt.plot(days, max_temp, label='max_temp')
plt.plot(days, min_temp, label='min_temp')
plt.plot(days, avg_temp, label='avg_temp')
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.legend()
```

```python
plt.plot(days, max_temp, label='max_temp')
plt.plot(days, min_temp, label='min_temp')
plt.plot(days, avg_temp, label='avg_temp')
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.legend(loc='upper right')
```

```python
plt.plot(days, max_temp, label='max_temp')
plt.plot(days, min_temp, label='min_temp')
plt.plot(days, avg_temp, label='avg_temp')
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.legend(loc='best')
```

```python
plt.plot(days, max_temp, label='max_temp')
plt.plot(days, min_temp, label='min_temp')
plt.plot(days, avg_temp, label='avg_temp')
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.legend(loc='best', shadow=True)
```

```python
plt.plot(days, max_temp, label='max_temp')
plt.plot(days, min_temp, label='min_temp')
plt.plot(days, avg_temp, label='avg_temp')
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.legend(loc='best', shadow=True, fontsize='large')
```

```python
plt.plot(days, max_temp, label='max_temp')
plt.plot(days, min_temp, label='min_temp')
plt.plot(days, avg_temp, label='avg_temp')
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.legend(loc='best', shadow=True)
plt.grid()
```
