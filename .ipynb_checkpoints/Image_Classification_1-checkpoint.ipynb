```python
def get_cropped_image_if_two_eyes(img_path):
    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >= 2:
            return roi_color
```

```python
import pywt
```

```python
def w2d(img, mode='haar', level=1):
    imArray = img
    #Datatype conversions
    #convert to grayscale
    imArray = cv2.cvtColor(imArray, cv2.COLOR_RGB2GRAY)
    #convert to float
    imArray =  np.float32(imArray)   
    imArray /= 255;
    # compute coefficients 
    coeffs = pywt.wavedec2(imArray, mode, level=level)

    #Process Coefficients
    coeffs_H = list(coeffs)  
    coeffs_H[0] *= 0;  

    # reconstruction
    imArray_H = pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H = np.uint8(imArray_H)

    return imArray_H
```

```python
im_haar = w2d(cropped_img_array,'db1',5)
plt.imshow(im_haar, cmap='gray')
```

```python
def get_cropped_image_if_two_eyes(img_path):
    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >= 2:
            return roi_color
```

```python
person_name_class_dict = {}
count = 0
for i in person_file_names_dict.keys():
    person_name_class_dict[i] = count
    count += 1
person_name_class_dict
```

```python
x, y = [], []
```

```python
for i, j in person_file_names_dict.items():
    for k in j:
        img = cv2.imread(k)
        if img is None:
            continue
        scaled_img = cv2.resize(img, (32, 32))
        img_haar = w2d(img, 'db1', 5)
        scaled_img_haar = cv2.resize(img_haar, (32, 32))
        combined_img = np.vstack((scaled_img.reshape(32 * 32 * 3, 1), scaled_img_haar.reshape(32 * 32, 1)))
        x.append(combined_img)
        y.append(person_name_class_dict)
```
