```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
```

```python
%matplotlib inline
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x, y, color='green', linewidth=5, linestyle='dotted')
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x, y, c='red', linewidth=5, linestyle='dashdot')
```
