```python
import pandas as pd
```

```python
df = pd.read_excel('survey.xls')
df
```

```python
pd.crosstab(df.Nationality, df.Handedness)
```

```python
pd.crosstab(df.Nationality, df.Sex)
```

```python
pd.crosstab(df.Nationality, [df.Handedness, df.Sex])
```

```python
pd.crosstab(df.Nationality, [df.Handedness, df.Sex], margins=True)
```

```python
pd.crosstab([df.Nationality, df.Sex], df.Handedness)
```

```python
pd.crosstab(df.Sex, df.Handedness, normalize='index')
```

```python
pd.crosstab(df.Sex, df.Handedness, normalize='columns')
```

```python
import numpy as np
```

```python
pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average)
```
