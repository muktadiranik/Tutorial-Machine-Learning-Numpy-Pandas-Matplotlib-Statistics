```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
```

```python
import numpy as np
```

```python
x_position = np.arange(len(company))
x_position
```

```python
plt.bar(x_position, revenue)
```

```python
plt.xticks(x_position, company)
plt.bar(x_position, revenue)
```

```python
plt.xticks(x_position, company)
plt.bar(x_position, revenue, label='revenue')
plt.xlabel('company')
plt.ylabel('revenue')
plt.title('stocks')
plt.legend()
```

```python
profit = [40, 2, 34, 12]
```
