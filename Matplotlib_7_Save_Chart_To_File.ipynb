```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car ", "Other Utilities"]
```

```python
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.2f%%', explode=[0.2, 0.1, 0.1, 0.1, 0.1], startangle=45)
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.2f%%', explode=[0.2, 0.1, 0.1, 0.1, 0.1], startangle=45)
plt.savefig('piechart.png')
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=2, autopct='%0.2f%%', explode=[0.2, 0.1, 0.1, 0.1, 0.1], startangle=45)
plt.savefig('piechart_2.png', bbox_inches='tight')
```
