```python
x = df_4.drop('HeartDisease', axis='columns')
```

```python
y = df_4.HeartDisease
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
```

```python
(xtrain, xtest, ytrain, ytest) = train_test_split(x_scaled, y, stratify=y, random_state=1, test_size=0.20)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
lr = LogisticRegression(max_iter=100000)
svc = SVC()
dtc = DecisionTreeClassifier()
rfc = RandomForestClassifier()
knn = KNeighborsClassifier()
gnb = GaussianNB()
```

```python
models = {
    'lr': {
        'model': LogisticRegression(max_iter=100000), 
        'params': {
            'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
        }
    }, 
    'svc': {
        'model': SVC(), 
        'params': {
            'kernel': ['linear', 'poly', 'rbf', 'sigmoid'], 
            'gamma': ['scale', 'auto']
        }
    }, 
    'dtc': {
        'model': DecisionTreeClassifier(), 
        'params': {
            'criterion': ['gini', 'entropy'], 
            'max_features': ['auto', 'sqrt', 'log2']
        }
    }, 
    'rfc': {
        'model': RandomForestClassifier(), 
        'params': {
            'criterion': ['gini', 'entropy'], 
            'max_features': ['auto', 'sqrt', 'log2']
        }
    }, 
    'knn': {
        'model': KNeighborsClassifier(), 
        'params': {
            'n_neighbors': [2, 3, 5], 
            'weights': ['uniform', 'distance'], 
            'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
        }
    }, 
    'gnb': {
        'model': GaussianNB(),  
        'params': {}
    }
}
```

```python
for i, j in models.items():
    print(i)
    print(j)
```

```python
scores = []
```

```python
for i, j in models.items():
    gscv = GridSearchCV(j['model'], j['params'], cv=5, return_train_score=False)
    gscv.fit(xtrain, ytrain)
    scores.append({
        'estimator_': i,  
        'best_score_': gscv.best_score_, 
        'best_params_': gscv.best_params_
    })
```

```python
scores
```

```python
df_scores = pd.DataFrame(scores, columns=['estimator_', 'best_score_', 'best_params_'])
```
