```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], color=['green', 'red'])
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], color=['green', 'red'], label=['men', 'women'])
plt.legend()
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], color=['green', 'red'], label=['men', 'women'])
plt.legend()
plt.xlabel('blood sugar')
plt.ylabel('number of patients')
plt.title('blood sugar analysis')
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], color=['green', 'red'], label=['men', 'women'], orientation='horizontal')
plt.legend()
plt.xlabel('blood sugar')
plt.ylabel('number of patients')
plt.title('blood sugar analysis')
```
